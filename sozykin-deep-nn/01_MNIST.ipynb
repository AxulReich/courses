# Распознавание рукописных цифр

```python
import numpy as np

# соответствует нейронной сети, слои которой соединены друг с другом
from keras.models import Sequential 
# в этом типе слоев все нейроны пред. уровня соединяются со всеми нейронами след. уровня
from keras.layers import Dense
# утилиты
from keras.utils import np_utils
# датасет MNIST http://yann.lecun.com/exdb/mnist/
from keras.datasets import mnist

# т.к. используем стохастические методы оптимизации, задаем seed для reproducibility
np.random.seed(1488)
```

## Load the data

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

## Preprocessing

преобразование размерности изображений

```python
X_train = X_train.reshape(60000, 28 * 28)
```

```python
X_test = X_test.reshape(10000, 28 * 28)
```

нормализация данных (интенсивность каждого пикселя приводится к значению от 0 до 1)

```python
X_train = X_train.astype('float32')

X_train /= 255
```

```python
X_test = X_test.astype('float32')

X_test /= 255
```

преобразуем метки в категории (из вектора длиной 60000 — в матрицу 60000×10, где 10 выходных сигналов по числу классов изображений)

```python
y_train
```

```python
y_train = np_utils.to_categorical(y_train, 10)
```

```python
y_train
```

```python
y_test = np_utils.to_categorical(y_test, 10)
```

## Создаём сеть

Создаем последовательную модель

```python
model = Sequential()
```

Добавляем входной слой сети

```python
model.add(Dense(800, # на входном слое 800 нейронов
                input_dim = 784, # у каждого нейрона 784 (28×28) входа
                init = 'normal', # веса входных значений инициализируются случайно, из норм. распр-я
                activation = 'relu'))# функция активации
```

Добавим скрытый слой, чтобы сеть была глубокой

```python
model.add( Dense(600, init = 'normal', activation = 'relu') )
```

Последний слой

```python
model.add( (Dense(10, init = 'normal', activation = 'softmax')) )
```

Компилируем модель (при компиляции мы задаем параметры, которые будут использоваться при обучении)

```python
model.compile(loss = 'categorical_crossentropy', # мера ошибки
              optimizer = 'SGD', # метод обучения — СГД
              metrics = ['accuracy']) # метрика, по которой будем выполнять оптимизацию
```

```python
model.summary()
```

## Обучаем сеть

```python
model.fit(X_train, y_train, # данные
          batch_size = 32, # размер мини-выборки для SGD
          nb_epoch = 10, # сколько раз обучаем сеть с помощью трейн сета
          verbose = 1)
```

## Оценим качество работы сети

```python
score = model.evaluate(X_test, y_test)
```

```python
score
```

Качество работы алгоритма

```python
print(round(score[1] * 100.0, 3))
```
